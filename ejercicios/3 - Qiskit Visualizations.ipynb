```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_state_city, plot_state_hinton, plot_state_paulivec
```

# 3 - Qiskit Visualization

## Exercise 1

Write down the usefulness of the following visualisation methods and describe when each one is particularly useful:

    - Plot_histogram
    - Plot_bloch_multivector
    - Coupling_Map

- `plot_histogram`

Permite visualizar de manera intuitiva los resultados de la ejecución de un circuito en un histograma, donde cada columna se corresponde con un resultado y su altura viene dada por los counts relativos a ese resultado.

Útil cuando queremos visualizar cómo se distribuyen los resultados.

- `plot_bloch_multivector`

Permite visualizar un estado cuántico en base a las esferas de bloch correspondientes a los qubits de ese estado.

Útil cuando queremos observar el estado de uno o varios qubits, pero no sirve cuando los qubits están entrelazados, ya que en ese caso los qubits individuales estarían en un estado mezcla (su estado no está bien definido).

- `coupling_map`

Permite visualizar la conectividad física entre los qubits del dispositivo cuántico o simuladore que estemos utilizando.

Útil para ver qué pares de qubits pueden interactua de manera directa a través de una puerta que opera sobre 2 qubits, como una `CNOT`.

## Exercise 2

Create a basic circuit consisting of two qubits, Hadamard gates, and S gates. Apply the following visualisation methods, commenting about the obtained results:

    - City plot
    - Hinton plot
    - Paulivec plot


Primero montamos el circuito

```python
qc = QuantumCircuit(2)

qc.h(range(2))
qc.s(0)

qc.draw(output="mpl")
```

- `plot_state_city`

Debido a las puertas Hadamard, las amplitudes son máximas, puesto que los 2 qubits están, por así decirlo, en el centro de la esfera de bloch.

Vemos que la mitad de las amplitudes son imaginarias y la otra mitad reales. Esto se debe a la rotación de $\frac{\pi}{2}$ realizada por la puerta S.

```python
plot_state_city(qc, figsize=(8, 8))
```

- `plot_state_hinton`

Esta visualización representa lo mismo que `plot_state_city` (los valores de la matriz densidad), pero en lugar de representar las amplitudes como "edificios", los representa como cuadrados, siendo blancos si son positivos y negros si son negativos. El área del cuadrado representa el valor de cada elemento de la matriz.

```python
plot_state_hinton(qc)
```

- `plot_state_paulivec(qc)`

Esta representación muestra la matriz de densidad utilizando los operadores de Pauli como la base. Cada barra muestra el valor esperado del operador de Pauli correspondiente.

En este caso, los coeficientes distintos de 0 son II, XI, YI y XY, todos con valor positivo. La ausencia de componentes en Z indica que el estado está en el plano X-Y de la esfera de Bloch.

```python
plot_state_paulivec(qc)
```
